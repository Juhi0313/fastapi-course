Neural Networks

```python
import torch
import torch.nn as nn ##nn = neural networks
import torch.nn.functional as F #functional is used for activation functions. these r the functions that r applied after the layers to break linearity.
import torch.optim as optim #optim for optimizers we'll use
from torch.utils.data import DataLoader, TensorDataset


from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
```

```python
X, y = load_breast_cancer(return_X_y=True) #X means info abt tumour and y means malignant or benign
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.2)

scaler = StandardScaler()

X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

```

```python
X_train_scaled
```

```python
y_train
```

```python
# now we'll turn this into tensors and then we want to use a data loader to easily load the data for our training process
X_train_scaled_tensor = torch.from_numpy(X_train_scaled).float()
X_test_scaled_tensor = torch.from_numpy(X_test_scaled).float()

y_train_tensor = torch.from_numpy(y_train).float().unsqueeze(1)
y_test_tensor = torch.from_numpy(y_test).float().unsqueeze(1)
```

```python
train_dataset = TensorDataset(X_train_scaled_tensor, y_train_tensor)
```

```python
X_train_scaled_tensor.shape
```

```python
# now we can turn this into loader
train_loader = DataLoader(train_dataset, batch_size=32, shuffle=True)
```

```python
#we'll create neural network as a class. this is the structure of neural networks
class BCNet(nn.Module):
  def __init__(self):
    super(BCNet, self).__init__()
    self.fc1 = nn.Linear(30, 64) #this is fully connnected dense linear layer. 64 is the neurons we want/outputs we want
    self.fc2 = nn.Linear(64, 32) #this layer takes output of previous layer as input
    self.fc3 = nn.Linear(32, 1) #1/single value is the actual output (0 or 1)

  #to put data in this structure:
  def forward(self, x): #x is the input
    #x = self.fcl(x) means we take this input vector/matrix if batches, we feed it through the network and get the result of taking the input and multiplying with weights of fully connected layer 1, adding the bias and getting some output.
    #in order to break the linearity:
    x = F.relu(self.fc1(x)) #relu is the activation function. this gives new value of x
    x = F.relu(self.fc2(x))
    x = torch.sigmoid(self.fc3(x)) #we used sigmoid because it produces value bw 0 and 1

    return x
    #this was the forward pass of our network

```

```python
model = BCNet()
```

```python
#since we r doing binary classification in this/Loss function:
criterion = nn.BCELoss() #bcloss = binary cross entropy loss
optimizer = optim.Adam(model.parameters(), lr=0.001) #lr = learning rate
```

```python
#now we'll define training loop/train the network
epochs = 20 #we r going to do 20 iterations on the same data

for epoch in range(epochs):
  model.train() #put the model in training model
  running_loss = 0.0

  #now we'll interate over our data loader to get individual batches
  for x_batch, y_batch in train_loader:
    #we r going to feed data thorugh our model, get prediction (which in beginning will be completley random), and then take true classification and calculate the loss (how wrong that model) and take that loss function and back-propogate (take derivative of) the loss and try to find minimum in that function (decent local minimum)
    optimizer.zero_grad() #remove gradients
    preds = model(x_batch)
    loss = criterion(preds, y_batch)
    #this was raw loss. now we need to back-propogate it

    loss.backward() #automatic differentiation
    optimizer.step() #update weights
    running_loss += loss.item() #keeping track of running loss aswell

  print(f'Epoch {epoch + 1}: Loss was {running_loss/len(train_loader)}')


```
